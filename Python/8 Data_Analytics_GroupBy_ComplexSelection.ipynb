# Data Analytics

## GroupBy and Complex Selection

```python
import pandas as pd
import numpy as np
```

```python
finance_info = pd.read_csv(r'tr_eikon_eod_data.csv') 
```

```python
finance_info = finance_info[:10]
```

```python
finance_info.head(15)
```

```python
finance_info['Letter'] = ['A', 'A', 'A', 'A', 'A','B', 'B', 'B', 'C', 'C']
```

```python
finance_info.head(14)
```

```python
finance_info.head(10)
```

```python
groups = finance_info.groupby('Letter')
```

nos agrupa toda las columnas, ahora podemos hacerle operaciones a las letras/grupos 

```python
groups.size()
```

```python
groups.mean()
```

```python
groups.max()
```

```python
groups.aggregate([min,max]).round(2)
```

```python
finance_info['Day'] = ['Weekday', 'Weekday', 'Weekend', 'Weekday', 'Weekend','Weekend', 'Weekday', 'Weekend',
                       'Weekend', 'Weekday']
```

```python
finance_info.head(10)
```

```python
groups = finance_info.groupby(['Letter','Day'])
```

```python
groups.size()
```

```python
groups.aggregate([sum,np.mean])
```

```python
groups[['AAPL.O','MSFT.O']].aggregate([sum,np.mean])
```
